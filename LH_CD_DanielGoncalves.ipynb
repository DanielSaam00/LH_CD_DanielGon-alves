```python
import seaborn as sns
import matplotlib.pyplot as plt


sns.barplot(data=df_filmes, x='Series_Title', y='IMDB_Rating', palette='viridis')
plt.title('Nota IMDb por Filme')
plt.xticks(rotation=45)
plt.show()


sns.barplot(data=df_filmes, x='Series_Title', y='No_of_Votes', palette='magma')
plt.title('Número de Votos por Filme')
plt.xticks(rotation=45)
plt.show()

sns.barplot(data=df_filmes, x='Series_Title', y='Meta_score', palette='coolwarm')
plt.title('Meta Score por Filme')
plt.xticks(rotation=45)
plt.show()
```

Os filmes selecionados para esta análise foram escolhidos com base em minha experiência pessoal como espectador, priorizando obras que, apesar de não serem amplamente populares, possuem características únicas e relevância artística. Mesmo aqueles com nota inferior a 7 no IMDb demonstram qualidades narrativas, visuais ou temáticas que os tornam marcantes.

A diversidade de gêneros - do terror experimental ao drama psicológico - revela como diferentes abordagens podem impactar o público de formas distintas. Além disso, muitos desses filmes influenciaram produções posteriores, seja pela estética, pela ousadia narrativa ou pela forma como abordam temas complexos.

A análise dos dados reforça que métricas como número de votos ou meta score não são os únicos indicadores de qualidade ou relevância. Filmes com menor visibilidade podem carregar grande valor cultural e emocional, especialmente quando observados fora da lógica comercial tradicional.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))
sns.histplot(df['IMDB_Rating'], bins=15, kde=True, color='skyblue')
plt.title('Distribuição das notas do IMDb')
plt.xlabel('Nota')
plt.ylabel('Frequência')
plt.show()

```

```python
df.groupby('Genre')['No_of_Votes'].mean().sort_values(ascending=False).plot(kind='bar', figsize=(10,5), color='orange')
plt.title('Média de votos por gênero')
plt.ylabel('Número médio de votos')
plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='Meta_score', y='IMDB_Rating', data=df, color='purple')
plt.title('Meta Score vs IMDb Rating')
plt.xlabel('Meta Score')
plt.ylabel('IMDb Rating')
plt.show()
```

```python
sns.regplot(x='Meta_score', y='IMDB_Rating', data=df, scatter_kws={'color': 'purple'}, line_kws={'color': 'black'})
plt.title('Meta Score vs IMDb Rating com linha de tendência')
plt.xlabel('Meta Score')
plt.ylabel('IMDb Rating')
plt.show()

```

A regressão entre Meta Score e IMDb Rating indica uma tendência positiva: filmes com maior avaliação da crítica tendem também a receber boas notas do público. Embora existam exceções, a linha de tendência sugere que há uma correlação moderada entre os dois indicadores. Isso reforça a ideia de que qualidade técnica e narrativa reconhecida por especialistas pode influenciar a recepção geral da audiência.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip().str.replace(' ', '_').str.lower()


df['meta_score'] = pd.to_numeric(df['meta_score'], errors='coerce')
df['imdb_rating'] = pd.to_numeric(df['imdb_rating'], errors='coerce')
df['no_of_votes'] = pd.to_numeric(df['no_of_votes'].astype(str).str.replace(',', ''), errors='coerce')


correlacoes = df[['meta_score', 'imdb_rating', 'no_of_votes']].corr()


plt.figure(figsize=(10, 8))
sns.heatmap(correlacoes, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Mapa de Correlação entre Variáveis')
plt.show()
```

```python
df['diferença_meta_imdb'] = df['meta_score'] - df['imdb_rating'] * 10
```

```python
df.sort_values('diferença_meta_imdb', ascending=False).head(10)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.scatterplot(data=df, x='meta_score', y='imdb_rating', hue='diferença_meta_imdb', palette='coolwarm')
plt.title('Diferença entre Meta Score e IMDb Rating')
plt.xlabel('Meta Score')
plt.ylabel('IMDb Rating')
plt.show()
```
